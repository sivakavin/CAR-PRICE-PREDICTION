```python
#importing required libraries
import selenium
import time
import pandas as pd
from selenium import webdriver

# Exception handling library
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException

# Require Data
Car = {}
Car["Brand"] = []
Car["Model"] = []
Car["Variant"] =[]
Car["Year"] = []
Car["driven kilometers"] =[]
Car["fuel"] = []
Car["Transmission"] = []
Car["number of owners"] =[] 
Car["location"] = []
Car["Price"] = []
```

```python
# Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\SIVA\DataScientiest\chromedriver.exe")
time.sleep(2)

# Opening olx website
url = "https://www.olx.in/"
driver.get(url)
time.sleep(2)
```

```python
#Navigating to car section
driver.find_element_by_xpath("//div[@class='_3AGJR _18NX_'][1]/a").click()
```

```python
run = True
while run:
    # if no content is avilable exception will be raised we need to handle this using try block
    try:
        for i in range(0,10):
            #loading full page
            driver.execute_script("window.scrollBy(0,10000)")
            time.sleep(2)
            driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']").click()
    except Exception:
        print("Fully loaded",Exception)
        run = False
```

```python
car_url = [] #empty list to store the each car URL
```

```python
url_car = driver.find_elements_by_xpath("//li[@class='EIR5N']/a")
time.sleep(1)
for i in url_car:
        car_url.append(i.get_attribute('href'))
```

```python
#Getting cardataset from the url

for i in set(car_url):
    
    driver.get(i)
    
    #Brand
    try:
        brand = driver.find_element_by_xpath("//*[@id='container']/main/div/div/div/div[1]/div/ol/li[7]/a").text.split('in')[0]
        Car['Brand'].append(brand)
    except:
        Car['Brand'].append("-")
    
    #Model
    try:
        model = driver.find_element_by_xpath("//*[@id='container']/main/div/div/div/div[1]/div/ol/li[8]/a").text.split('in')[0]
        Car['Model'].append(model)
    except:
        Car['Model'].append("-")
        
    #Variant
    try:
        variant = driver.find_element_by_xpath("//div[@class='_3Mfu-']/div[3]").text
        Car['Variant'].append(variant)
    except:
        Car['Variant'].append("-")
        
    #Year
    try:
        year = driver.find_element_by_xpath("//div[@class='_3Mfu-']/div[2]").text.split('(')[-1].replace(')','')
        Car['Year'].append(year)
    except:
        Car['Year'].append("-")
        
    #fuel
    try:
        fuel = driver.find_element_by_xpath("//div[@class='_1u27F'][1]/div").text
        Car['fuel'].append(fuel)
    except:
        Car['fuel'].append("-")

    #transmission
    try:
        transmission = driver.find_element_by_xpath("//div[@class='_1u27F'][3]/div").text
        Car['Transmission'].append(transmission)
    except:
        Car['Transmission'].append("-")

    #km
    try:
        km = driver.find_element_by_xpath("//div[@class='_1u27F'][2]/div").text
        Car['driven kilometers'].append(km)
    except:
        Car['driven kilometers'].append("-")
    
    #owner
    try:
        owner = driver.find_element_by_xpath("//div[1][@class='_1ThNy']/div[2]/div[2]").text
        Car['number of owners'].append(owner)
    except:
        Car['number of owners'].append("-")
    #location
    try:
        loc = driver.find_element_by_xpath("//div[2][@class='_1ThNy']/div[2]/div[2]").text.split(',')[-1]
        Car['location'].append(loc)
    except:
        Car['location'].append("-")
        
    #price
    try:
        price = driver.find_element_by_xpath("//div[@class='_3FkyT']").text.split('₹')[-1].strip()
        Car['Price'].append(price)
    except:
        Car['Price'].append("-")
```

```python
Car_Data = pd.DataFrame(Car)
Car_Data.head()
```

```python
Car_Data.to_csv(r"C:\Users\SIVA\DataScientiest\Car_Data.csv",na_rep='-')
```
